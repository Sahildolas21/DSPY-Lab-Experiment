
**Subarrays as no-copy views**

One important–and extremely useful–thing to know about array slices is that they return views rather than copies of the array data. This is one area in which NumPy array slicing differs from Python list slicing: in lists, slices will be copies. Consider our two-dimensional array from before:


````python
#Extract 2*2 sub array from x2
x2_sub = x2[:2, :2]
print(x2_sub)
````

````python
x2_sub[0, 0] = 99
#Above statement not oly modifies subarray but also the original array as well
print(x2_sub)
````

Despite the nice features of array views, it is sometimes useful to instead explicitly copy the data within an array or a subarray. This can be most easily done with the copy() method:

````python
import numpy as np

x2 = np.random.randint(10, size=(3, 4))
x2_sub = x2[:2, :2]        # this is a view
x2_sub_copy = x2[:2, :2].copy()  # this is an actual copy

x2_sub_copy[0, 0] = 42

print(x2_sub_copy)
print(x2)
````

**Introducing UFuncs**

For many types of operations, NumPy provides a convenient interface into just this kind of statically typed, compiled routine. This is known as a vectorized operation. This can be accomplished by simply performing an operation on the array, which will then be applied to each element. This vectorized approach is designed to push the loop into the compiled layer that underlies NumPy, leading to much faster execution

````python
#Consider following loop based implementation to ind reciprocals for each element of an array
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
````

````python
#Same operatoin using UFuncs applying '/' over array element
1.0 / big_array
````



Vectorized operations in NumPy are implemented via ufuncs, whose main purpose is to quickly execute repeated operations on values in NumPy arrays. UFuncs are extremely flexible – before we saw an operation between a scalar and an array, but we can also operate between two arrays as well as multidimentional arrays.

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented using Python loops, especially as the arrays grow in size. Any time you see such a loop in a Python script, you should consider whether it can be replaced with a vectorized expression.

**4 ) Consider array object named Obj with dimentions 3*4*2 and you want to convert it to 4*6 shape exlpain how will you do it ? Show this using a code cell, taking an example of 3*4*2 np array of random int between 1 and 50.**

**Answer :**

To convert an array with dimensions 3x4x2 into a 4x6 shape, you must first ensure the total number of elements remains the same in both configurations.

a ) The original array has \(3\times 4\times 2=24\) elements.

b ) The target shape has \(4\times 6=24\) elements.

Because the total number of elements matches, the conversion is possible by using a combination of flattening (or 'unraveling') and then reshaping the array.

**The general process involves:**

a ) Flatten the original multi-dimensional array into a one-dimensional array (vector) containing all 24 elements in sequence.

b ) Reshape this one-dimensional array into the desired 4x6 structure.

The following Python code cell demonstrates this process using the numpy library, as it is the standard for numerical operations and multi-dimensional arrays in Python

````python
import numpy as np

# 1. Create the initial 3*4*2 array with random integers between 1 and 50
#    The total number of elements is 3 * 4 * 2 = 24
original_shape = (3, 4, 2)
# Setting a seed for reproducibility of random numbers
np.random.seed(42)
original_array = np.random.randint(1, 51, size=original_shape)

print(f"Original array shape: {original_array.shape}")
print("Original array:\n", original_array)
print("-" * 30)

# 2. Reshape the array to the target 4*6 shape
#    The total number of elements (24) remains the same.
target_shape = (4, 6)
# NumPy's reshape method handles the flattening and reshaping in one step
reshaped_array = original_array.reshape(target_shape)

print(f"Reshaped array shape: {reshaped_array.shape}")
print("Reshaped array:\n", reshaped_array)
print("-" * 30)

# Verification: Check if both arrays contain the same elements in the same order
# when flattened, demonstrating the underlying data is preserved.
are_elements_equal = np.array_equal(original_array.flatten(), reshaped_array.flatten())
print(f"Are the flattened elements of both arrays equal? {are_elements_equal}")
````

**7 ) If you were to filter sub array view from above Obj object and want to apply modification in original and/or sub obj change respective element in other how will you achieve it?**

**Answer :**

o filter a subset (view) from a Python object (like a list of dictionaries) and ensure modifications to the subset reflect in the original, and vice versa, you must create a new container holding
references to the original mutable objects rather than copying them.

Using a list comprehension to filter for mutable objects (like dictionaries) creates a new list, but the items within that list still point to the same memory addresses as the original list.


````python
# 1. Original data (list of dictionaries - mutable)
original_data = [
    {'id': 1, 'status': 'active'},
    {'id': 2, 'status': 'inactive'},
    {'id': 3, 'status': 'active'}
]

# 2. Filter a sub-array view (list of references)
# Only dictionaries with status 'active'
sub_view = [item for item in original_data if item['status'] == 'active']

print("Original before:", original_data)
# [{'id': 1, 'status': 'active'}, {'id': 2, 'status': 'inactive'}, {'id': 3, 'status': 'active'}]

# 3. Modify sub_view -> Changes in original_data
sub_view[0]['status'] = 'updated'

# 4. Modify original_data -> Changes in sub_view
original_data[2]['status'] = 'updated'

print("Sub-view:", sub_view)
# [{'id': 1, 'status': 'updated'}, {'id': 3, 'status': 'updated'}]

print("Original after:", original_data)
# [{'id': 1, 'status': 'updated'}, {'id': 2, 'status': 'inactive'}, {'id': 3, 'status': 'updated'}]
````
