# **Handling Missing Values**

Different strategies are applied:

Numerical columns → Fill with mean/median
Categorical columns → Fill with mode
Drop columns if too many missing values

```python
import pandas as pd

df = pd.read_csv("customer data.csv")

# Fill numeric columns with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill categorical columns with mode
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

print(df.isnull().sum())
```

# **Data Cleaning**

Data cleaning involves:

Removing duplicate records
Converting categorical data into numerical format

```python
import pandas as pd

# Load dataset
df = pd.read_csv("customer data.csv")

# Remove duplicates
df.drop_duplicates(inplace=True)

# Handle missing values
df.fillna(df.mean(numeric_only=True), inplace=True)
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

# Convert column names to lowercase
df.columns = df.columns.str.lower()

# Strip extra spaces from string values
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()

print("Cleaned Data:\n", df.head())
```
