**1.What is need for Hypothesis Testing ?**

**Answer :**

* **The Need for Hypothesis Testing**

Hypothesis testing is a fundamental tool in statistics because it provides a structured way to make decisions or draw conclusions about a population basedon sample data. Here’s why it is needed:

* **Dealing with Uncertainty**

Data collected from samples always contains variability. Hypothesis testing helps us determine whether observed differences or effects are due to random chance or represent real, significant patterns in the population.

* **Objective Decision-Making**

Instead of relying on intuition or subjective judgment, hypothesis testing uses probability theory to provide a formal framework for deciding whether to accept or reject assumptions (hypotheses).

* **Comparing Groups or Conditions**

It allows us to test claims such as:

"Does a new drug improve recovery rates compared to the standard treatment?"

"Is the average exam score of one class higher than another?" These comparisons are essential in science, business, medicine, and social research.

* **Controlling Error Rates**

Hypothesis testing incorporates concepts like p-values and significance levels (α) to control the risk of making incorrect conclusions (Type I and Type II errors).

* **Foundation for Statistical Inference**

It bridges the gap between descriptive statistics (summarizing data) and inferential statistics(drawing conclusions about populations). Without hypothesis testing, we cannot generalize findings beyond the data we directly observe

**3.What is a student t distribution?**

**Answer :**

**Student’s t-distribution**

The Student’s t-distribution is a continuous probability distribution that is widely used in statistics, particularly in hypothesis testing and confidence interval estimation. It is applied when the sample size is relatively small and the population standard deviation is unknown.Compared to the normal distribution, the t-distribution has heavier tails, which allows it to
accommodate the greater variability and uncertainty inherent in small samples. As the sample size increases, the t-distribution gradually approaches the normal distribution. It forms the basis for calculating the t-statistic and corresponding p-values, which are used to determine whether there is enough evidence to reject a null hypothesis about a population mean.

**4.What is level of significance ? What role t plays in hypothesis Testing**

**Answer :**

Level of Significance and Role of t in Hypothesis Testing

The level of significance, denoted by α (alpha), is the probability threshold used in hypothesis testing to decide whether to reject the null hypothesis. It represents the maximum risk of making a Type I error, which occurs when the null hypothesis is wrongly rejected even though it is true. Commonly used significance levels are 0.05, 0.01, and 0.10.

The t-statistic plays a central role in hypothesis testing when the population standard deviation is unknown and the sample size is small. It measures how far the sample mean deviates from the hypothesized population mean in terms of standard errors. This t-value is then compared against the critical value from the Student’s t-distribution corresponding to the chosen level of significance and degrees of freedom. If the t-statistic falls in the critical region, the null hypothesis is rejected; otherwise, it is not rejected. Thus, the t-statistic provides a quantitative basis for making decisions in hypothesis testing.


**5.Design a hypothesis testing for identifying average age of students in TE IT class is 19 or not.**

**Answer :**

Hypothesis Testing: Average Age of Students in TE IT Class
Objective: To determine whether the average age of students in the TE IT class is 19 years.

**Step 1: Define Hypotheses**

* Null Hypothesis (H₀): The average age of students in the TE IT class is 19 years. (μ = 19)
* Alternative Hypothesis (H₁): The average age of students in the TE IT class is not 19
years. (μ ≠ 19)

**Step 2: Select Significance Level**

* Choose a level of significance (α), typically 0.05.

**Step 3: Collect Sample Data**

* Collect a random sample of students’ ages from the TE IT class.

**Step 4: Calculate Test Statistic**

* Use the t-test since the population standard deviation is unknown and the sample size is
likely small.
* Formula for t-statistic:
[ t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}} ]

Where:
  * ( \bar{x} ) = sample mean

  * ( \mu_0 ) = hypothesized mean (19)

  * ( s ) = sample standard deviation

  * ( n ) = sample size

**Step 5: Determine Critical Value / p-value**

* Compare the calculated t-statistic with the critical t-value from the Student’s tdistribution table at α level and df = n-1.
* Alternatively, compute the p-value.

**Step 6: Make Decision**

* If |t-statistic| > t-critical or p-value < α → Reject H₀ (average age is not 19).
* Otherwise → Fail to reject H₀ (insufficient evidence to conclude the average age differs from 19).

**Step 7: Conclusion**

* State the result in context: whether the data provides enough evidence to support that the average age of students is 19 or not.

**3. Python Implementation**

Here’s a ready-to-run Python snippet for your notebook

```python
import numpy as np
from scipy import stats

# Sample data: ages of TE IT students
ages = np.array([19, 20, 18, 19, 20, 19, 18, 19, 20, 19])

# Hypothesized mean
mu_0 = 19

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(ages, mu_0)

# Significance level
alpha = 0.05

print("Sample mean:", np.mean(ages))
print("t-statistic:", t_stat)
print("p-value:", p_value)

# Decision based on p-value
if p_value > alpha:
    print("Fail to reject H0: There is not enough evidence to say the average age is different from 19")
else:
    print("Reject H0: The average age is significantly different from 19")
```

**4. How to Interpret**

* p-value tells the probability of observing your sample mean if H₀ is true.
* Compare p-value with 0.05 → decide whether to accept or reject H₀.
