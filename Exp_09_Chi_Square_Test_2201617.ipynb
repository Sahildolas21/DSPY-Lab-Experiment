**1.When do we go for Chi square test ?**

**Answer :**

The Chi-square test is a non-parametric statistical test used to analyze
categorical data and examine whether there is a significant association between
variables or whether observed data fits an expected distribution.

We go for a Chi-square test in the following situations:
1. When the data is categorical in nature (such as gender, class, category, yes/no
responses, etc.), not numerical.
2. When we want to test the independence between two categorical variables.
This is known as the Chi-square test of independence. It helps determine
whether the variables are related or occur independently.
3. When we want to compare observed frequencies with expected frequencies.
This is known as the Chi-square goodness-of-fit test. It checks whether the
observed distribution matches a theoretical or expected distribution.
4. When the sample size is sufficiently large, and expected frequencies in each
category are generally 5 or more, ensuring validity of the test.
5. When the observations are independent of each other, meaning one
observation does not influence another.
6. When the data is in the form of counts or frequencies, not percentages,
proportions, or continuous measurements.

In Python, the Chi-square test is typically performed using functions like
scipy.stats.chi2_contingency() for testing independence and
scipy.stats.chisquare() for goodness-of-fit. These functions compute the Chisquare statistic and p-value, which are used to decide whether to reject or accept the null hypothesis.

If the p-value is less than the chosen significance level (such as 0.05), we reject the null hypothesis, indicating that there is a significant relationship or difference. If the p-value is greater, we fail to reject the null hypothesis.

To perform the Chi-square test for independence, the following steps are followed:
1. State the hypotheses: The null hypothesis (H₀) states that the two variables are
independent. The alternative hypothesis (H₁) states that the two variables are
dependent (there is an association between them).
2. Create a contingency table: The data is organized into a table showing the
frequency distribution of the variables.
3. Calculate expected frequencies: Expected frequency for each cell is calculated
using the formula: Expected = (Row Total × Column Total) / Grand Total
4. Compute the Chi-square statistic: The formula is: χ² = Σ [(Observed −
Expected)² / Expected] This measures the difference between observed and
expected frequencies.
5. Determine degrees of freedom: Degrees of freedom = (number of rows − 1) ×
(number of columns − 1)
6. Find the p-value: Using the Chi-square distribution, the p-value is obtained
based on the computed statistic and degrees of freedom.
7. Make a decision: If the p-value is less than the significance level (e.g., 0.05), we
reject the null hypothesis and conclude that the variables are dependent. If the
p-value is greater than the significance level, we fail to reject the null hypothesis
and conclude that the variables are independent.

**3.Explain role of Critical value and p-value in Chi square test.**

**Answer :**

In the Chi-square test, both the critical value and the p-value are used to
make decisions about the null hypothesis, but they follow two different approaches.

**Role of Critical Value:**

The critical value is a threshold obtained from the Chi-square distribution table based on the chosen level of significance (α, such as 0.05) and the degrees of freedom.

After calculating the Chi-square test statistic (χ²), it is compared with the critical value:

If χ² (calculated) > χ² (critical), we reject the null hypothesis.

If χ² (calculated) ≤ χ² (critical), we fail to reject the null hypothesis.

Thus, the critical value defines the boundary of the rejection region. If the test statistic falls in this region, it indicates that the observed data significantly differs from the expected data.

**Role of p-value:**

The p-value represents the probability of obtaining the observed results (or more extreme) assuming that the null hypothesis is true.

* If p-value ≤ α (e.g., 0.05), we reject the null hypothesis.

* If p-value > α, we fail to reject the null hypothesis.

The p-value provides a direct measure of evidence against the null hypothesis. A
smaller p-value indicates stronger evidence against H₀.

**Summary:**

Both methods lead to the same conclusion:

* The critical value approach compares the test statistic with a threshold.

* The p-value approach compares probability with the significance level.

In practice, the p-value method is more commonly used in Python (e.g., using
scipy.stats.chi2_contingency() ), as it directly provides the probability for
decision-making.

**5.What is degrees of freedom and how to calculate it while performing chi square test.**

**Answer :**

**Degrees of Freedom (DoF)** represent the number of independent values that are
free to vary when calculating a statistical measure under certain constraints.
In the Chi-square test, degrees of freedom determine the shape of the Chi-square
distribution and are essential for finding the critical value or p-value.

**Calculation of Degrees of Freedom in Chi-square Test:**

1. **For Chi-square Test of Independence:** When data is arranged in a
contingency table: Degrees of freedom = (number of rows − 1) × (number of
columns − 1)

2. **For Chi-square Goodness-of-Fit Test:** Degrees of freedom = (number of
categories − 1 − number of estimated parameters)

**Explanation:**

* In a contingency table, once the row totals and column totals are fixed, not all
cell values are independent. Some values are automatically determined, which
reduces the freedom of variation.

* Degrees of freedom account for these constraints and adjust the test
accordingly.

**Importance:**

* It is used to determine the correct Chi-square distribution.

* It helps in calculating the critical value and p-value.

* Incorrect degrees of freedom can lead to wrong conclusions in hypothesis
testing.
